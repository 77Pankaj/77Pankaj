```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# Training of Model
model1  = LogisticRegression()
model1.fit(X_train,Y_train)
```

# Our best model which gives the required accuracy score which is more than 80%.

```python
model_predictions = model1.predict(X_train)
Score = metrics.accuracy_score(model_predictions,Y_train)
print("Accuracy Score for model1 is:",Score) # it is quite good
```

```python
model2 = XGBClassifier() # This creates the case of Overfitting
model2.fit(X_train,Y_train)
model_predictions1 = model2.predict(X_train)
print("Accuracy Score for model2 is:",metrics.accuracy_score(model_predictions1,Y_train))
```

```python
model3 = RandomForestClassifier()
model3.fit(X_train,Y_train)
```

```python
model_predictions2 = model3.predict(X_train) # Also a case of overfitting
print("Accuracy Score for model3 is:",metrics.accuracy_score(model_predictions2,Y_train))
```

```python
model4  = SVC()
model4.fit(X_train,Y_train)
```
